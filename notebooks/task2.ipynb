```python
import pandas as pd
import logging
```

```python
logging.info('load the data')
df = pd.read_csv('../data/data.csv')
print(df.head(5))
```

```python
logging.info('overview of the data')
print(df.shape)
print(df.dtypes)
print(df.info())
```

```python
logging.info('summary statistics')
print(df.describe())
```

```python
logging.info('missing values')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
logging.info('Histogram plot for the Amount')
plt.figure(figsize=(10, 8))
sns.histplot(df['Amount'], bins=20, kde=True)
plt.title('Histogram of Amount')
plt.xlabel("Amount")
plt.ylabel('Frequency')
plt.show()
```

```python
logging.info('Desnisty plot for the values')
plt.figure(figsize=(10,8))
sns.kdeplot(df['Value'])
plt.title('Density plot of Value')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
logging.info('Bar plot for the category')
plt.figure(figsize=(10,8))
sns.countplot(data=df,x='ProductCategory')
plt.title('Bar plot of Poroduct Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
logging.info('Bar plot for the FraudResult')
plt.figure(figsize=(10,8))
sns.countplot(data=df,x='FraudResult')
plt.title('Bar plot of Fraud Result')
plt.xlabel('Fraud Result')
plt.ylabel('Count')
plt.show()
```

```python
logging.info('Correlation plot')

numerical_values = df.select_dtypes(include=['float64', 'int64'])

corr_mat = numerical_values.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_mat, annot=True, cmap='coolwarm')
plt.title('Correlation plot')
plt.show()
```

```python
logging.info('check for the missing values')
missing_values = df.isnull().sum()
print(missing_values)
```

```python
logging.info('Visualize the missing values')
plt.figure(figsize=(10, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='binary', linewidths=0.5)
plt.title('Missing Values Heatmap')
plt.show()
```

```python
logging.info('outlier detection')
plt.figure(figsize=(10,8))
sns.boxplot(x=df['Amount'])
plt.title('Box plot of Amount')
plt.xlabel('Amount')
plt.show()
```

```python
logging.info('outlier detection using Value')
plt.figure(figsize=(10,8))
sns.boxplot(x=df['Value'])
plt.title('Box plot of Value')
plt.xlabel('Value')
plt.show()
```
