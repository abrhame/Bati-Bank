```python
import pandas as pd
import logging
```

```python
logging.info('load the data')
df = pd.read_csv('../data/data.csv')
print(df.head(5))
```

```python
logging.info('overview of the data')
print(df.shape)
print(df.dtypes)
print(df.info())
```

```python
logging.info('summary statistics')
print(df.describe())
```

```python
logging.info('missing values')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
logging.info('Histogram plot for the Amount')
plt.figure(figsize=(10, 8))
sns.histplot(df['Amount'], bins=20, kde=True)
plt.title('Histogram of Amount')
plt.xlabel("Amount")
plt.ylabel('Frequency')
plt.show()
```

```python
logging.info('Desnisty plot for the values')
plt.figure(figsize=(10,8))
sns.kdeplot(df['Value'])
plt.title('Density plot of Value')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
